**1-Scatter Plots**

```
# prepare some data
ages = df['Age']
fare = df['Fare']
# output to static HTML file
output_file("bokeh_scatter_example.html", title="Relationship between Fare and Age for each Passengers")

# create a new plot with a title and axis labels
fig = figure(title="Relationship between Fare and Age for each Passengers", x_axis_label='Age', 
              y_axis_label='Fare')

# add a Scatter plot
fig.circle(ages, fare, size=4, alpha=0.5)
#show result
show(fig)
```

**Bar Chart**

```
from bokeh.plotting import figure, output_file, show

output_file('vbar.html')

p = figure(plot_width=400, plot_height=400)
p.vbar(x=df['Pclass'], width=0.5, bottom=0,
       top=[1.2, 2.5, 3.7], color="firebrick")

show(p)
```

**line plot**

```
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, output_file, show
# output to static HTML file
output_file("vline_stack.html")
# prepare some data
source = ColumnDataSource(data=dict(
    x=[0,1],
    y1=df['Survived'],
   
))
p = figure(plot_width=400, plot_height=400)
# add a line plot
p.vline_stack(['y1', 'y2'], x='x', source=source)

# show the results
show(p)
```
