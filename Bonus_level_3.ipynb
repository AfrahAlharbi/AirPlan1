** Create a chart to tell me something about this data**

```
from bokeh.io import output_notebook, show
output_notebook()
from bokeh.plotting import figure, output_file, show
```

```
import pandas as pd
df= pd.read_parquet('https://github.com/gumdropsteve/datasets/raw/master/airlines.parquet')
df.tail(100)
```

line plot for Distance 

```
x = [i for i in range(100)]
y = [i for i in df['Distance'].tail(100)]

# output to static HTML file
output_file("lines.html")

# create a new plot with a title and axis labels
p = figure(title="simple line example", x_axis_label='x', y_axis_label='y')

# add a line renderer with legend and line thickness
p.line(x, y, legend_label="Distance.", line_width=2)

# show the results
show(p)
```
