```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

###Data Source
The data used in this project is Titanic data. This data is from Kaggle and you can find the data description in  this [**link**](https://www.kaggle.com/c/titanic/data).


```
df = pd.read_csv('https://raw.githubusercontent.com/gumdropsteve/intro_to_python/main/day_09/data/titanic.csv')
df
```

### Bar charts:

```
fig, ax = plt.subplots(1, 2, figsize = (10, 8)) 
df["Embarked"].value_counts().plot.bar(color=['Gold' ,'YellowGreen','LightCoral'], ax = ax[0])
ax[0].set_ylabel("Passengers",fontweight='bold',fontsize=10)
ax[0].set_xlabel("Embarked",fontweight='bold',fontsize=10)
ax[0].set_title("Number Of Passengers By Embarked",fontweight='bold',fontsize=12)

color2=['LightCoral' ,'Gold','YellowGreen']
sns.countplot("Embarked", hue = "Survived", data = df, ax = ax[1],palette=color2)
ax[1].set_ylabel("Passengers",fontweight='bold',fontsize=10)
ax[1].set_xlabel("Embarked",fontweight='bold',fontsize=10)
ax[1].set_title("Embarked: Survived vs Dead",fontweight='bold',fontsize=12)

plt.show();
```

This plot shows us that Embarkation Southampton has the maximum number of passengers so it has the maximum number of passengers not survived and survived. 
While Embarkation  Cherbourg and Queenstown are less.

### Scatter charts:

```
# Change color style
plt.style.use("default")

# Create scatter chart comparing mpg and horsepower
df.plot(
    kind="scatter", 
    x="Age", 
    y="Fare",
    color='LightCoral',
    figsize=(15,8),
    alpha=0.5,
    s=80
    )

# Add title
plt.title("Relationship between Age and Fare");
```

The chart shows that most passengers pay less than 100, while a few riders pay more than 100, also most of passengers have the age between 20 to 40.

```
# LinePlot for Age & Fare
sns.set_theme(style="white")
sns.lineplot(x="Age", y="Fare", 
            sizes=(50, 400), alpha=0.7, color='YellowGreen',
             data=df).set_title("Fare amount among Age",fontweight='bold',fontsize=12)
 
 
```

Despite Age, most passengers did not have a total fare amount of more than 100.
